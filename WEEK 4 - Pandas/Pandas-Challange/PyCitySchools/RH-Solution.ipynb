```python
import pandas as pd
import numpy as np
```

```python
schools_data_path = "../../04-Pandas/Instructions/PyCitySchools/Resources/schools_complete.csv"
students_data_path = "../../04-Pandas/Instructions/PyCitySchools/Resources/students_complete.csv"
schools_data = pd.read_csv(schools_data_path)
students_data = pd.read_csv(students_data_path)
school_data_complete = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])
```

```python


# District Summary

#Create a high level snapshot (in table form) of the district's key metrics, including:

#Total Schools
total_schools = school_data_complete.groupby(["school_name"])

#Total Students
total_students = schools_data["size"].sum()

#Total Budget
total_budget = schools_data["budget"].sum()
 
#Average Math Score
avg_math_score= (students_data["math_score"].sum())/(students_data["math_score"].count())

#Average Reading Score
avg_reading_score = (students_data["reading_score"].sum())/(students_data["reading_score"].count())

math = [x for x in students_data["math_score"] if x >= 70]
reading = [x for x in students_data["reading_score"] if x >= 70]
passing_math = pd.DataFrame({
    "math":math})
#% Passing Math
perc_passing_math = passing_math["math"].count()/ students_data["math_score"].count()

#% Passing Reading
passing_reading = pd.DataFrame({
    "reading":reading})
perc_passing_reading = passing_reading["reading"].count()/ students_data["reading_score"].count()

#Overall Passing Rate (Average of the above two)
overral = (perc_passing_math + perc_passing_reading)/2

district_summary = pd.DataFrame({
    "Total Schools": [total_schools], 
    "Total Students":[total_students],
    "Total Budget":[total_budget],
    "Average Math Score":[avg_math_score],
    "Average Reading Score":[avg_reading_score],
    "% Passing Math":[perc_passing_math],
    "% Passing Reading":[perc_passing_reading],
    "Overral Passing Rate":[overral],
    
})

district_summary.style.format({"Total Budget": "${:,.2f}", 
                       "Average Reading Score": "{:.1f}", 
                       "Average Math Score": "{:.1f}", 
                       "% Passing Math": "{:.1%}", 
                       "% Passing Reading": "{:.1%}", 
                       "Overral Passing Rate": "{:.1%}"})

```

```python
#School Summary

#Create an overview table that summarizes key metrics about each school, including:
school_data_complete["reading_pass_indicator"] = [1 if x >= 70 else 0 for x in school_data_complete["reading_score"]]
school_data_complete["math_pass_indicator"] = [1 if x >= 70 else 0 for x in school_data_complete["math_score"]]

#School Name
by_school_name = school_data_complete.set_index("school_name").groupby("school_name")
school_display = schools_data.set_index("school_name")


#School Type
school_type = school_data_complete.set_index("school_name").groupby("type")

#Total Students
total_students = by_school_name["Student ID"].count()

#Total School Budget

total_budget = schools_data.set_index("school_name")["budget"]


#Per Student Budget
per_student_budget = schools_data.set_index("school_name")["budget"]/ schools_data.set_index("school_name")["size"]

#Average Math Score
avg_math = by_school_name["math_score"].mean()
avg_reading = by_school_name["reading_score"].mean()



#% Passing Math

passing_math = by_school_name["math_pass_indicator"].sum()
perc_math = passing_math / total_students


#% Passing Reading

passing_reading = by_school_name["reading_pass_indicator"].sum()
perc_reading = passing_reading / total_students

#Overall Passing Rate (Average of the above two)

overall = (perc_math + perc_reading)/2

school_summary = pd.DataFrame({
    "Total Students": total_students,
    "Per Student Budget": per_student_budget,
    "Total School Budget": total_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_reading,
    '% Passing Math': perc_math,
    '% Passing Reading': perc_reading,
    "Overall Passing Rate": overall
})



school_summary.head()
```
